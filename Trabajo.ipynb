## Ejemplo del algoritmo

El algoritmo tiene como objetivo, representar todos los recorridos posibles, desde un vertice de partida y representarlos en forma de árbol.

```python
class Node:
    data = None
    children = {}

    def __init__(self, data = None):
        self.data = data

    def __str__(self):
        return str(self.get_data())

    def __bool__(self):
        return self.data is not None

    def get_child(self, index = 0):
        if(index in self.children.keys()):
            return self.children[index]
        else:
            raise IndexError(f'Child {index} not defined')

    def get_data(self):
        return self.data

    def set_data(self, data):
        self.data = data

    def add_child(self, data):
        self.children[data] = Node(data)


class BinaryTreeNode(Node):
    def __init__(self, data = None):
        self.data = data
        self.children = {}

    def side(self, side: str):
        if(side in ['left','right','l','r']):
            return int(side in ['right','r'])
        else:
            raise ReferenceError(f'Unknown value {side}')

    def add_leaf(self, data, side: str):
        self.children[self.side(side)] = BinaryTreeNode(data)

class Graph:
    def __init__(self):
        self.vertices = {}

    def __iter__(self):
        return  iter(self.vertices.values())

    def add_vertex(self, vertex: Node):
        if vertex.data not in self.vertices:
            self.vertices[vertex.data] = vertex
            return True
        else:
            return False

    def add_edge(self, vertex1: Node, vertex2: Node):
        if vertex1.data in self.vertices and vertex2.data in self.vertices:
            self.vertices[vertex1.data].add_child(vertex2)

    def get_vertices(self):
        return self.vertices.keys()
```

```python
g = Graph()

vertices = list()

# Crea vertices
for i in range(15):
    vertices.append(Node(i))

# Agrega vertices a grafo
for v in vertices:
    g.add_vertex(v)
```
